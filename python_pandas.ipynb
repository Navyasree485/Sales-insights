# Sales transactions data containing duplicates and wrong data 

```python
# import pandas as pd
df = pd.read_csv('sales_transactions.csv')

# Deleting the rows containing sales amount either 0 or negative
for x in df.index:
    if df.loc[x, "sales_amount"] <= 0:
        df.drop(x, inplace = True)
# print(df.to_string())

for x in df.index:
    if df.loc[x, "currency"] == "INR" or df.loc[x, "currency"] == "USD":
        df.drop(x, inplace = True)


# Changing the amount from USD to INR
for x in df.index:
     if df.loc[x, "currency"] == "USD\r" :
        df.loc[x, "sales_amount"] = x*75
        df.loc[x, "currency"] = "INR\r"
print(df.to_string())



```
